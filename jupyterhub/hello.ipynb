```python
# Import tensorflow first
# This is a Windows based file, might not work on *nix environments
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(7)
```

```python
dataset = np.loadtxt('indian-diabetes.csv', delimiter=',')
x = dataset[:, 0:8]
y = dataset[:, 8:]
```

```python
# Create the model
model = Sequential()
```

```python
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(x, y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
loss = plt.plot(history.epoch[2:], history.history['loss'][2:], color='r', label='Loss')
acc = plt.plot(history.epoch[2:], history.history['acc'][2:], color='b', label='Accuracy')
plt.legend(['Model loss', 'Model accuracy'])
plt.show()
```
